```python
import sys
import os
from importlib import reload
import pylab as plt
import skimage
import numpy as np
from skimage import exposure

sys.path.insert(0, '../src')
import deepleeo.dataset.image_utils as iutils
import deepleeo.utils.geofunctions as gf

reload(iutils)
```

```python
DATA_DIR = "/home/raian/doutorado/Dados/2015/LC082250642015081601T1-SC20180823104500"
raster_bands = [os.path.join(DATA_DIR, "LC08_L1TP_225064_20150816_20170406_01_T1_sr_band3.tif"),
                os.path.join(DATA_DIR, "LC08_L1TP_225064_20150816_20170406_01_T1_sr_band4.tif"),
                os.path.join(DATA_DIR, "LC08_L1TP_225064_20150816_20170406_01_T1_sr_band5.tif"),
                os.path.join(DATA_DIR, "LC08_L1TP_225064_20150816_20170406_01_T1_sr_band6.tif"),
                os.path.join(DATA_DIR, "LC08_L1TP_225064_20150816_20170406_01_T1_sr_band7.tif")]

band_names = ["Green", "Red", "NIR", "SWIR_1", "SWIR_2"]
out_img = os.path.join(DATA_DIR, "../stacked_imgs", "LC08_L1TP_225064_16082015.tif")
```

```python
iutils.stack_bands(raster_bands, out_img, band_names)
```

```python
raster_img = gf.load_image(out_img, no_data=-9999)
plt.figure(figsize=(10,10))
plt.title('RGB Composition')
raster_img = skimage.img_as_ubyte(raster_img)
for i in range(raster_img.shape[2]):
    p2, p98 = np.percentile(raster_img[:,:,i], (2, 98))
    raster_img[:,:,i] = exposure.rescale_intensity(raster_img[:,:,i], in_range=(p2, p98))
plt.imshow(raster_img[:,:,[3,2,1]])
plt.axis('off')
```
